```python

#Class is a movie, it maps out its genres, and links to past and future movies in that genre
class Movie:
    def __init__(self,title,date,genres):
        self.title=title
        self.date=date
        self.genres=genres
        self.futuregenresfilms= [None] * len(genres)
        self.pastgenresfilms=[None]*len(genres)
    #Returns location in array that genre is
    def FindGenre(self,genre):

        for x in range(len(self.genres)):
            if self.genres[x]==genre:

                return x
        
        
        return None
    #Returns future film in specific genre
    def ReturnFutureGenreFilm(self,genre):
        return self.futuregenresfilms[self.FindGenre(genre)]
    #Same but with past
    def ReturnPastGenreFilm(self,genre):
        return self.pastgenresfilms[self.FindGenre(genre)]
    #Assigns a movie to be the past film in that genre in the pastgenresfilms array
    def AssignFilmToPastGenreFilms(self,genre,film):
        self.pastgenresfilms[self.FindGenre(genre)]=film
    def AssignFilmToFutureGenreFilms(self,genre,film):
        self.futuregenresfilms[self.FindGenre(genre)]=film
    #Checks to see if the movie is older than a certain date
    def AmIOlder(self,Date):
        return (self.date>Date)
        
#Actor with list of genres as a dictionary        
class Actor:
    def __init__(self,FirstName,LastName):
        self.FirstName=FirstName
        self.LastName=LastName
        self.genres={}
        
    #Add movie to dictionary of movies
    def AddMovie(self,Movie):
        #assigns the film to each genre
        for genre in Movie.genres:
            #Tests to see if the genre is in the dictionary, if not it adds it as the first movie
            try:
                genretrail=self.genres[genre]
                if (genretrail.AmIOlder(Movie.date)):
                    try:

                        genretrail.pastgenresfilms[genretrail.FindGenre(genre)]=Movie

                        Movie.futuregenresfilms[Movie.FindGenre(genre)]=genretrail

                        self.genres[genre]=Movie
                    except Exception as e:
                        print (1)
                        print (e)
                #If Film is older than the oldest currently in the list assign it to the last spot in the list and assign it as the future film of the current last, and set current last as the past in that genre
                elif (~genretrail.AmIOlder(Movie.date) and genretrail.futuregenresfilms[genretrail.FindGenre(genre)]==None):
                    try:
                        genretrail.futuregenresfilms[genretrail.FindGenre(genre)]=Movie
                        Movie.pastgenresfilms[Movie.FindGenre(genre)]=genretrail
                    except Exception as e:
                        print (2)
                        print (e)
                #If its not last in the list keep going till its either younger than the next film or the last in the list and add it
                #Have to assign future movie to current movie future list, current movie to future movie past list
                #Past movie to past movie list of current movie, and assign current movie to past movie future movie list
                
                else:
                    while (~genretrail.AmIOlder(Movie.date) and genretrail.futuregenresfilms[genretrail.FindGenre(genre)]!=None):
                        if (genretrail.futuregenresfilms[genretrail.FindGenre(genre)].AmIOlder(Movie.date)):
                            try:
                                temp=genretrail.futuregenresfilms[genretrail.FindGenre(genre)]
                                Movie.futuregenresfilms[Movie.FindGenre(genre)]=temp
                                Movie.pastgenresfilms[Movie.FindGenre(genre)]=genretrail
                                genretrail.futuregenresfilms[genretrail.FindGenre(genre)].pastgenresfilms[genretrail.futuregenresfilms[genretrail.FindGenre(genre)].FindGenre(genre)]=Movie
                                genretrail.futuregenresfilms[genretrail.FindGenre(genre)]=Movie
                            except Exception as e:
                                print (3)
                                print (e)
                        else:
                            genretrail=genretrail.futuregenresfilms[genretrail.FindGenre(genre)]
                            break
                    if (~genretrail.AmIOlder(Movie.date) and genretrail.futuregenresfilms[genretrail.FindGenre(genre)]==None):
                        print('F')
                        try:
                            Movie.pastgenresfilms[Movie.FindGenre(genre)]=genretrail
                            genretrail.futuregenresfilms[genretrail.FindGenre(genre)]=Movie
                        except Exception as e:
                            print (4)
                            print (e)
            except Exception as e:
                print (10)
                print ('New Genre Added')
                print (e)
                self.genres[genre]=Movie
    #Print out the timeline of a genre
    def GetMovieTimeline(self,genre,movies=[]):
        temp=self.genres[genre]
        movies.append((temp.title,temp.date))
        while temp.futuregenresfilms[temp.FindGenre(genre)]!=None:
            #print (temp.title)
            temp=temp.futuregenresfilms[temp.FindGenre(genre)]
            movies.append((temp.title,temp.date))
        return movies
                
            
```
