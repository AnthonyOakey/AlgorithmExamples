```python
#Tree using two seperate classes to be able to change Node and have fuller control
#Double Linked the true using parent in order to be able to go backwards
class Node:
    def __init__(self,value):
        self.value=value
        self.parent=None
        self.left=None
        self.right=None

class Tree:
    def __init__(self):
        self.root=None
    #Inserts Node
    def insert(self,Node,root=None):
        if self.root==None:
            self.root=Node
        else:
            if root==None:
                root=self.root
            #Basic insert, Bigger go right and if there is no Value set it as that node
            #If Node is added it sets its parent to current node
            if Node.value>root.value:
                if root.right==None:
                    Node.parent=root
                    root.right=Node
                else:
                    self.insert(Node,root.right)
            elif Node.value<root.value:
                if root.left==None:
                    Node.parent=root
                    root.left=Node
                else:
                    self.insert(Node,root.left)
    #Find lowest value from spot used to rebalance
    def lowestvalue(self,root=None):
        if root==None:
            root=self.root
        while root.left:
            root=root.left
        return root
    #Find largest just in case it is needed
    def largestvalue(self,root=None):
        if root==None:
            root=self.root
        while root.right:
            root=root.right
        return root        
    #Replace a child in the parent    
    def replacechild(self,oldnode,newnode):
        if oldnode.parent.right==oldnode:
            oldnode.parent.right=newnode
        elif oldnode.parent.left==oldnode:
            oldnode.parent.left=newnode
    #Remove the node
    def remove(self,value,root=None):
        if root==None:
            root=self.root
        node=self.FindValue(value,root)
        if node:
            if node.left and node.right:
                temp=self.lowestvalue(root.right)
                temp.parent.left=None
                self.replacechild(node,temp)
                temp.parent=node.parent
                temp.left=node.left
                temp.left.parent=temp
                temp.right=node.right
                temp.right.parent=temp
                node.value=None
                node=None
                
                #self.remove(temp.value,node.right)
            elif node.left:
                temp=node.left
                self.replacechild(node,temp)
                node=None
            elif node.right:
                temp=node.right
                self.replacechild(node,temp)
                node=None
            else:
                self.replacechild(node,None)
            print ('{} Removed'.format(value))
        else:
            print ("{} Not Found".format(value))
    #Get the center of the array just makes it easier to read if a function is used
    def getmiddle(array):
        return array[int(len(array)/2)]
    #It will insert into the tree a list in a balanced way
    def insertBalanceList(self,root,lis):
        if len(lis)<4:
            if len(lis)==3:
                mid=int(len(lis)/2)
                temp=Node(lis[mid])
                
                lis.remove(lis[mid])
                for item in lis:
                    self.insert(Node(item),temp)
                self.insert(temp,root)
            else:
                for item in lis:
                    self.insert(Node(item),root)                
        else:
            mid=int(len(lis)/2)
            temp=Node(lis[mid])
            self.insert(temp,root)
            self.insertBalanceList(temp,lis[:mid+1])
            self.insertBalanceList(temp,lis[mid-1:])
            
            
    #Rebalanced the tree by gathering all the nodes and then inserting them in a balanced way
    def Rebalance(self,root=None):
        if root==None:
            root=self.root
        values=self.BFS(root,[])
        values.append(root.value)
        values.sort()
        if root.parent:
            self.replacechild(root,None)
            self.insertBalanceList(root.parent,values)
        else:
            self.root=None
            self.insertBalanceList(self.root,values)
        
            
            
        
            
            
    def BFS(self,root=None,queue=[]):
        if root==None:
            root=self.root
            queue.append(root.value)

        if root.left!=None:
            queue.append(root.left.value)
        if root.right!=None:
            queue.append(root.right.value)
        if root.left!=None:
            queue=self.BFS(root.left,queue)
        if root.right!=None:    
            queue=self.BFS(root.right,queue)
        return queue
    def DFS(self,root=None,queue=[]):
        if root==None:
            root=self.root
            queue.append(root.value)

        if root.left!=None:
            queue=self.BFS(root.left,queue)
        if root.right!=None:
            queue=self.BFS(root.right,queue)
        return queue
    #Find and return a value    
    def FindValue(self,value,root=None):
        if root==None:
            root=self.root
        if root.value==value:
            print ("Value {} found".format(value))
            return root
        elif value>root.value:
            if root.right==None:
                print ("Value {} NOT Found".format(value))
            else:
                node=self.FindValue(value,root.right)
                if node:
                    return node
        else:
            if root.left==None:
                print ("Value {} NOT Found".format(value))
            else:
                node=self.FindValue(value,root.left)
                if node:
                    return node
    #Clear The Tree
    def Clear(self):
        self.root=None
        self.queue=[]
```
